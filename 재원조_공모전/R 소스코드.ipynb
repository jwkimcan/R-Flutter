## 쇼핑몰 별 상위 검색어 300위 분석

### 쇼핑몰 통합

```r
# 상위 검색어 300위 파일 불러오기
all2=read.csv('/Users/dasom/Desktop/data_story/all_s_300.csv')
```

```r
# 상위 검색어 10위 추출하기
colnames(all2)=c('index','word','count')
all3=all2%>%head(10)%>%arrange(count) # 상위 10위 검색어를 검색빈도가 높은 순서대로 정렬
all3
```

```r
# 상위 검색어 10위 시각화하기
options(warn = -1)
par(family = "AppleGothic")
barplot(all3$count,
        main='4대 쇼핑몰 통합 상위 10위 검색어',
       xlim=c(0,35000),
       names=all3$word,
        las=1,
        col='darkblue',
        cex.names=0.6,
      horiz=T)
```

### 옥션

```r
# 상위 검색어 300위 파일 불러오기
ac2=read.csv('/Users/dasom/Desktop/data_story/auction_s_300.csv')
```

```r
# 상위 검색어 10위 추출하기
colnames(ac2) = c('word','count')
ac3=ac2%>%head(10)%>%arrange(count) # 상위 10위 검색어를 검색빈도가 높은 순서대로 정렬
ac3
```

```r
# 상위 검색어 10위 시각화하기
options(warn = -1)
par(family = "AppleGothic")
    barplot(head(ac3$count,10),
        main='옥션 상위 10위 검색어',
       xlim=c(0,4000),
       names=head(ac3$word,10),
        las=1,
        col='purple',
        cex.names=0.7,
      horiz=T
       
       )
x=0.7
for(i in 1:10){
    text((ac3[i,2])+400,x,(ac3[i,2]),cex=0.8)
    x=x+1.2
}
```
