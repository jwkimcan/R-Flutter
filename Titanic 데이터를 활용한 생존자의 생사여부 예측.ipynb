## 티켓 등급 분석

```r
# Pclass의 분포
Pclass_1 <- nrow(subset(titanic, titanic$Pclass == 1)) ; Pclass_1
Pclass_2 <- nrow(subset(titanic, titanic$Pclass == 2)) ; Pclass_2
Pclass_3 <- nrow(subset(titanic, titanic$Pclass == 3)) ; Pclass_3

table(titanic$Pclass)
```

```r
# 1등 객실의 여성 생존율
Pclass_1_female <- nrow(subset(titanic, titanic$Sex == 'female' & titanic$Pclass == 1))
Pclass_1_female
Pclass_1_female_sur <- nrow(subset(titanic, titanic$Survived == 1 & titanic$Sex == 'female' & titanic$Pclass == 1))
Pclass_1_female_sur
Pclass_1_female_sur_per <- ((Pclass_1_female_sur/Pclass_1_female)*100)
Pclass_1_female_sur_per
```

```r
cat('1등 객실의 여성 생존율은', Pclass_1_female_sur, '/', Pclass_1_female, ',',round(Pclass_1_female_sur_per,2), '%')
```

```r
# 1,2,3등 객실의 여성 생존율을 for문을 통해 해결

Pclass_female <- c()
Pclass_female_sur <- c()
Pclass_female_sur_per <- c()

for(i in 1:length(unique(titanic$Pclass))){
    
    Pclass_female <- 
    nrow(subset(titanic, titanic$Sex == 'female' & titanic$Pclass == i))
    
    Pclass_female_sur <- 
    nrow(subset(titanic, titanic$Survived == 1 & titanic$Sex == 'female' & titanic$Pclass == i))
    
    Pclass_female_sur_per <- 
    ((Pclass_female_sur/Pclass_female)*100)
    
    cat(i,'등 객실의 여성 생존율은', Pclass_female_sur, '/', Pclass_female, ',',
        round(Pclass_female_sur_per,2), '%\n')
}
```

```r
# 1,2,3등 객실의 남성과 여성 생존율을 이중 for문을 통해 해결

Pclass_person <- c()
Pclass_person_sur <- c()
Pclass_person_sur_per <- c()

for(j in c('female','male')){
    
    for(i in 1:length(unique(titanic$Pclass))){
    
        Pclass_person <- 
        nrow(subset(titanic, titanic$Sex == j & titanic$Pclass == i))

        Pclass_person_sur <- 
        nrow(subset(titanic, titanic$Survived == 1 & titanic$Sex == j & titanic$Pclass == i))

        Pclass_person_sur_per <- 
        ((Pclass_person_sur/Pclass_person)*100)

        cat(i,'등 객실의', ifelse(j == 'female','여성','남성') ,'생존율은', 
            Pclass_person_sur, '/', Pclass_person, ',',
            round(Pclass_person_sur_per,2), '%\n')
    }
    cat('--------------------------------------------\n')
}
```

```r
# 1,2,3등 객실의 남성과 여성 생존율을 벡터로 담아내기

maleData <- c()
femaleData <- c()

for(j in c('female','male')){
    
    for(i in 1:length(unique(titanic$Pclass))){
    
        Pclass_person <- 
        nrow(subset(titanic, titanic$Sex == j & titanic$Pclass == i))

        Pclass_person_sur <- 
        nrow(subset(titanic, titanic$Survived == 1 & titanic$Sex == j & titanic$Pclass == i))

        Pclass_person_sur_per <- 
        ((Pclass_person_sur/Pclass_person)*100)

        cat(i,'등 객실의', ifelse(j == 'female','여성','남성') ,'생존율은', 
            Pclass_person_sur, '/', Pclass_person, ',',
            round(Pclass_person_sur_per,2), '%\n')
        
        if(j == 'female'){
            femaleData[i] = round(Pclass_person_sur_per,2)
        }else{
            maleData[i] = round(Pclass_person_sur_per,2)
        }
    }
    cat('--------------------------------------------\n')
}
```

## RandomForest 작업

```r
# Sampling 
library(caret)
set.seed(1234)
samp <- createDataPartition(train_df_T$Survived, p = 0.7, list = F)
samp
```

```r
#학습용 
data.train <- train_df_T[samp,]
#검즘용
data.test <- train_df_T[-samp,]
```

```r
#학습용 재구성 
train.data = data.train[target_col]
```

```r
#검증용 재구성 
test.data = data.test[target_col]
```

```r
library(randomForest)
```

```r
# 학습시키기 
rf= randomForest(Survived ~ . , data= train.data , ntree = 100)
rf%>%plot
```

```r
#문제와 정답 만들기 
x = subset(test.data,select = -Survived)
y = test.data$Survived
```

```r
#예측하기 
pred = predict(rf,x)
pred
```

```r

pred[pred <=0.5]=0
pred[pred>0.5]=1
mean(pred==y)
```

```r
table(pred,y)
```
