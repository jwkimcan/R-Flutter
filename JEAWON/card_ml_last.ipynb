```r
library(caret)
library(randomForest)
```

```r
# 0,1,2 3경우의수 똑같은 비중을 위해 sample
card0=card %>% filter(credit==0)   
card1=card %>% filter(credit==1) %>%  sample_n(3222)
card2=card %>% filter(credit==2) %>%  sample_n(3222)
```

```r
cardf =card0 %>% rbind(card1) %>% rbind(card2)   
```

```r
cardf
```

```r
cardf
```

```r
cardf %>% colnames
```

```r
card$credit=card$credit %>% as.factor
```

```r
set.seed(1382)
samp = createDataPartition(cardf$credit,p = 0.7,list=F)
samp
```

```r
#학습용
data.train = card[samp,]
#검증용
data.test = card[-samp,]
```

```r
target_col = c('child_num','DAYS_BIRTH','DAYS_EMPLOYED',
               'work_phone','phone','email','begin_month','credit','gendercode','carcode',
               'realitycode','income_total1','income_total2','income_total3','income_total4',
               'income_total5','income_total6','income_total7','income_total8','income_type1',
               'income_type2','income_type3','income_type4','income_type5','edu_type1','edu_type2',
               'edu_type3','edu_type4','edu_type5','family_type1','family_type2','family_type3',
               'family_type4','family_type5','house_type1','house_type2','house_type3',
               'house_type4','house_type5','house_type6')
train.data = data.train[target_col]
test.data = data.train[target_col]
rf = randomForest(credit~.,data=train.data,ntree = 1600,mtry = 9,integer=T)
x = subset(test.data,select = -credit)
y = test.data$credit
pred = predict(rf,x)
```

```r
mean(pred==y)
```
