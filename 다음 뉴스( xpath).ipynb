# 뉴스30개 발췌하기

```r
web=GET("https://news.daum.net/ranking/kkomkkom")
web
```

```r
html =htmlParse(web)
html
```

### xpath로 뉴스 제목 발췌하기

```r
# xpath확인 
xpathSApply(html,'//*[@id="mArticle"]/div[2]/ul[3]/li[1]/div[2]/strong/a',
           xmlValue
           )
```

```r
## 반복문으로 뉴스 제목 발췌하기 
for (i in 1:30){
    str_xpath = sprintf('//*[@id="mArticle"]/div[2]/ul[3]/li[%i]/div[2]/strong/a',i)
    str_01 = xpathSApply(html,str_xpath,xmlValue)
    cat("No :",i,",제목 :",str_01,"\n")
}
```

```r
## dataFrame으로 저장하기 
content = c() 
count = 0
for (i in 1:30){
    count=count+1
    str_xpath = sprintf('//*[@id="mArticle"]/div[2]/ul[3]/li[%i]/div[2]/strong/a',i)
    str_01 = xpathSApply(html,str_xpath,xmlValue)
    content[count]=str_01
}
daum_df_title = data.frame(순위=1:30,기사제목=content)
daum_df_title
colnames(daum_df_title)=c('rank','title')
```

```r
## csv로저장하기 
write.csv(daum_df_title,'./daum_df.csv',row.names=F)
```
